```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
sX = np.array([1532, 1688, 1773, 1800])
sY = np.array([0.67598, 0.39722, 0.27874, 0.22623])

cX = np.array([1250, 1565, 1730, 1800])
cY = np.array([0.82787, 0.41520, 0.28429, 0.22368])

plt.plot(sX, sY, label="DC shunt motor", marker=".")
plt.plot(cX, cY, label="DC cumulative motor", marker=".")
plt.legend(bbox_to_anchor=(0.5, 1.1), ncol=2, mode="expand")
plt.show(True)
```
