```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
x_a = np.array([])
y_a = np.array([])

for x in range(10, 0, -1):
    y = x**2 + 3*x - 2
    x_a = np.append(x_a, x)
    y_a = np.append(y_a, y)

x_head = 0
y_head = x_head**2 + 3*x_head - 2
x_tail = 10
y_tail = x_tail**2 + 3*x_tail - 2
dx = x_head - x_tail
dy = y_head - y_tail

plt.plot(x_a,y_a)
plt.arrow(x_head, y_head, dx, dy)
plt.show()
```

```python
x_a = np.array([])
y_a = np.array([])

for x in range(100, 0, -1):
    y = math.sqrt(3 - 5*x + x**2 + x**3) / x - 1
    x_a = np.append(x_a, x)
    y_a = np.append(y_a, y)

plt.plot(x_a,y_a)
plt.show()

print(x_a)
print(y_a)
```
