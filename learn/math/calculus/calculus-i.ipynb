```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x_a = np.array([])
y_a = np.array([])

x = 0
while x < 9:
    x_a = np.append(x_a, x)
    y_a = np.append(y_a, math.sin(x))
    x += 0.001

plt.plot(x_a,y_a)
```
