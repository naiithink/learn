# Limits

> Getting closer and closer.

```
let f(x) = ( ((3) - (5 * x) + (x ** 2) + (x ** 3)) ** 0.5 ) / (x - 1)

determine what happens to f(x) when x approachs 1 from the right.
```

````python
def f(x):
    res = ( ((3) - (5 * x) + (x ** 2) + (x ** 3)) ** 0.5 ) / (x - 1)
    return res

x = np.array([1.1, 1.01, 1.001, 1.0001, 1.00001])
fx = np.array([])

for xi in x:
    fx = np.append(fx, f(xi))

plt.plot(x,fx)
plt.show()
````
